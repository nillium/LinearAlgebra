```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(8, 3), sharey=True)


units_untrans = np.array([[1,0], [0,1]])
units_trans = np.array([[1,0], [1,1]])
origin = np.array([[0,0], [0,0]])

axs[0].text(1.1, 0.1, '$\mathbf{\hat{i}}$', fontsize = 16, color = 'r')
axs[0].text(0.1, 1.1, '$\mathbf{\hat{j}}$', fontsize = 16, color = 'g')
axs[0].set_title("Untransformed")
axs[1].set_title("Transformed")
axs[2].set_title("Untransformed")
for items in axs:
    items.set_xlabel("X-axis", fontsize = 12)
    items.set_ylabel("Y-axis", fontsize = 12)
    items.grid()
    items.set_box_aspect(1)
    items.set_xlim([-2, 2])
    items.set_ylim([-2, 2])

axs[0].quiver(*origin, units_untrans[:,0], units_untrans[:,1], angles='xy', scale_units='xy', scale=1, color=['r', 'g'], zorder=3)
axs[1].quiver(*origin, units_trans[:,0], units_trans[:,1], angles='xy', scale_units='xy', scale=1, color=['r', 'g'], zorder=3)


plt.show()
```
