```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])

units = np.array([[1,0], [0,1]])
origin = np.array([[0,0], [0,0]])
ax.quiver(*origin, units[:,0], units[:,1], angles='xy', scale_units='xy', scale=1, color=['r', 'g'], zorder=3)

plt.text(1.1, 0.1, '$\mathbf{\hat{i}}$', fontsize = 12, color = 'r')
plt.text(0.1, 1.1, '$\mathbf{\hat{j}}$', fontsize = 12, color = 'g')

plt.xlabel("X-axis", fontsize = 12)
plt.ylabel("Y-axis",fontsize = 12)
plt.grid()
plt.show()
```
